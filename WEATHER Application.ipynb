```python

# WEATHER APPlication

import tkinter as tk
from tkinter import messagebox
import random

# Dummy weather database
WEATHER_DATA = {
    "Hyderabad": ("Clear Sky", 32, 55, 3.5),
    "Bangalore": ("Cloudy", 27, 65, 4.2),
    "Chennai": ("Sunny", 34, 60, 5.1),
    "Delhi": ("Haze", 30, 48, 2.8),
    "Mumbai": ("Rainy", 29, 78, 6.0),
    "Kolkata": ("Humid", 31, 70, 3.9)
}

def get_weather():
    city = entry_city.get().strip()

    if not city:
        messagebox.showerror("Error", "Enter a city name")
        return

    if city not in WEATHER_DATA:
        messagebox.showerror("Error", "City not found in demo database")
        return

    condition, temp_c, humidity, wind = WEATHER_DATA[city]

    if unit_var.get() == "F":
        temp = round((temp_c * 9/5) + 32, 2)
        symbol = "°F"
    else:
        temp = temp_c
        symbol = "°C"

    result.set(
        f"City: {city}\n"
        f"Temperature: {temp} {symbol}\n"
        f"Humidity: {humidity}%\n"
        f"Wind Speed: {wind} m/s\n"
        f"Condition: {condition}"
    )

try:
    root.destroy()
except:
    pass

root = tk.Tk()
root.title("Basic Weather App")

tk.Label(root, text="Enter City").grid(row=0, column=0)
entry_city = tk.Entry(root)
entry_city.grid(row=0, column=1)

unit_var = tk.StringVar(value="C")
tk.Radiobutton(root, text="Celsius", variable=unit_var, value="C").grid(row=1, column=0)
tk.Radiobutton(root, text="Fahrenheit", variable=unit_var, value="F").grid(row=1, column=1)

tk.Button(root, text="Get Weather", command=get_weather).grid(row=2, column=0, columnspan=2)

result = tk.StringVar()
tk.Label(root, textvariable=result, justify="left").grid(row=3, column=0, columnspan=2)

root.mainloop()
```
